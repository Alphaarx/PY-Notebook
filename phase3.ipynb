## Phase 3: Data Analysis and Visualization

## Sales data of Business

```python
# Data structure:
# [restaurant id, 2021, 2022,2023, 2024]
sales_data = np.array([
[1, 150000,180000, 220000, 250000],  # Paradise Biryani
[2,120000, 148800,160000,190000],    # Beijing Bites
[3,200000, 230000,260000, 300000],   # Pizza Hub
[4,180000, 210000, 240000, 270000],  # Burger Point
[5, 160000, 185000,205000, 230000]   # Chai Point
])  


print("=== Sales Data ===")
print("sales_data: ", sales_data.shape)
print("sales_data", sales_data[:,1:])  # [:(take all rows) , 1: (take allcolumns from 1 to end)]
print("sales_data", sales_data[1:,1:])  # [1: (take all rows from 1 to end) , 1: (take all columns from 1 to end)]
```

# sale per year


```python
#total sales
print("=== Total Sales Per Year ===")
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
print("total_sales_per_year:", total_sales_per_year)

#minimum sales
print("=== Minimum Sales Per Year ===")
minimum_sales_per_year = np.min(sales_data[:, 1:], axis=1) #min give minimum value in each column or row
print("minimum_sales_per_year:", minimum_sales_per_year)

#maximum sales
print("=== Maximum Sales Per Year ===")
maximum_sales_per_year = np.max(sales_data[:, 1:], axis=0) #max give maximum value in each column or row
print("maximum_sales_per_year:", maximum_sales_per_year)

#average sales
print("=== Average Sales Per Year ===")
average_sales_per_year = np.mean(sales_data[:, 1:], axis=1) #mean give average value in each column or row
print("average_sales_per_year:", average_sales_per_year)
```

```python
cumsum= np.cumsum(sales_data[:, 1:], axis=1) #cumsum give cumulative sum value in each column or row
print("=== Cumulative Sales Per Year ===")
print(cumsum)
plt.Figure(figsize=(10, 5))
years = np.array([2021, 2022, 2023, 2024])  # years for x-axis
plt.plot(np.mean(cumsum,axis=0),years, marker='*') # plot x

plt.xlabel('Cumulative Sales')
plt.ylabel('Year')
plt.show()
```

```python
 #vector operation
vector1=np.array([1,2,3,4,5,6])
vector2=np.array([7,8,9,10,11,12])

print("Vector addition: ",vector1+vector2)
print("Vector multiplication: ",vector1*vector2)
print("Dot product of two vectors: \n", np.dot(vector1, vector2))  # dot product of two vectors

#matrix multiplication
matrix1=np.array([[1,2,3],[4,5,6]])
matrix2=np.array([[7,8],[9,10],[11,12]])
print("Matrix multiplication: \n", np.dot(matrix1, matrix2)) # dot product of two matrices

angel=np.arccos(np.dot(vector1, vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2))) # angle between two vectors in radians
print("Angle between two vectors: ", angel) # angle between two vectors in 
```
